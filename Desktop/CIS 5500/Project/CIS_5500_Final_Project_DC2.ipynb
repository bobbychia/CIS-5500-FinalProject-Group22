```python
# MAP RULE:
# N:删除
# KG:00
# PK:000
# UG:13

edu.drop(edu[edu['school_level'] == 'N'].index, inplace=True)
edu.drop(edu[edu['lowest_grade'] == 'N'].index, inplace=True)
edu.drop(edu[edu['highest_grade'] == 'N'].index, inplace=True)

edu['lowest_grade'] = edu['lowest_grade'].replace('KG', '00')
edu['lowest_grade'] = edu['lowest_grade'].replace('PK', '000')
edu['highest_grade'] = edu['highest_grade'].replace('KG', '00')
edu['highest_grade'] = edu['highest_grade'].replace('PK', '000')
edu['lowest_grade'] = edu['lowest_grade'].replace('UG', '13')
edu['highest_grade'] = edu['highest_grade'].replace('UG', '13')
```

```python
print(edu['school_level'].unique())
print(edu['lowest_grade'].unique())
print(edu['highest_grade'].unique())
```

```python
#清理完之后有 99882
# len(edu)
```

```python
# Phone Number Cleaning
edu['phone_number'].apply(lambda x: isinstance(x, str)).sum()
```

```python
print(edu['school_level'].value_counts(dropna=False))
print(edu['lowest_grade'].value_counts(dropna=False))
print(edu['highest_grade'].value_counts(dropna=False))
```

```python
# zip_code 转成字符串 school_id 转成字符串
edu['zip_code'] = edu['zip_code'].astype(str)
edu['school_id'] = edu['school_id'].astype(str)
```

```python
######## Real Estate Data Cleaning ########
```

```python
re = pd.read_csv('Real_Estate_Final.csv')
re
```

```python
re.dtypes
```

```python
re["city"].value_counts(dropna=False)
```

```python
# Transfer the zip_code and broker_id into string type for better handling of categorical data
re['zip_code'] = re['zip_code'].astype(int).astype(str)
re['broker_id'] = re['broker_id'].astype(int).astype(str)
```

```python
# Bedrooms and Bathrooms should be integers, but there are some float values. Let's check how many there are.
re[['bedrooms', 'bathrooms', 'street_address','house_size','price']].apply(lambda x: (x % 1 != 0).sum())
```

```python
re["bedrooms"] = re['bedrooms'].astype(int)
re["bathrooms"] = re['bathrooms'].astype(int)
re["street_address"] = re['street_address'].astype(int)
re["house_size"] = re['house_size'].astype(int)
re["price"] = re['price'].astype(int)
```

```python
# 处理日期
re["previous_sold_date"] = pd.to_datetime(re["previous_sold_date"], errors='coerce')
```

```python
# City 大写
re["city"] = re["city"].str.upper()
```

```python
# 删除 street_address 
re.drop(columns=['street_address'], inplace=True)
```

```python
re["property_id"] = re["property_id"].astype(str)
```

```python
# listing_status 
print(re['listing_status'].value_counts(dropna=False))

# map rule:
# 0: sold
# 1: for_sale

re['listing_status'] = re['listing_status'].map({'sold': 0, 'for_sale': 1})

re['listing_status'].astype(int)
```

```python
########### IRS Data Cleaning ###########
```

```python
irs = pd.read_csv('IRS_Final.csv')
irs
```

```python
irs.dtypes
```

```python
float_cols = ['num_returns', 'total_income', 'wage_income', 'interest_income', 'dividend_income', 'capital_gain']

for col in float_cols:
    has_decimal = (irs[col] % 1 != 0).sum()
    print(f"{col}: {has_decimal} 个有小数")
```

```python
for col in float_cols:
    irs[col] = irs[col].fillna(0).astype(int)
```

```python
# zipcode 转成字符串
irs['zip_code'] = irs['zip_code'].astype(int).astype(str)
```
