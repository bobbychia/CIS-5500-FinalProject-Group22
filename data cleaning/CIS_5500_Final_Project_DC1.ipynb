```python
education_rename = {
    'ncessch': 'school_id',
    'schnam09': 'school_name',
    'mstree09': 'street_address',
    'mcity09': 'city',
    'mstate09': 'state',
    'mzip09': 'zip_code',
    'member09': 'enrollment',
    'phone09': 'phone_number',
    'ulocal09': 'locale_code',
    'type09': 'school_type',
    'level09': 'school_level',
    'gslo09': 'lowest_grade',
    'gshi09': 'highest_grade',
    'status09': 'school_status'
}

Education = Education.rename(columns=education_rename)
Education.head()
```

```python
irs_rename = {
    'STATE': 'state',
    'zipcode': 'zip_code',
    'agi_stub': 'income_bracket',
    'N1': 'num_returns',
    'A00100': 'total_income',
    'A00200': 'wage_income',
    'A00300': 'interest_income',
    'A00600': 'dividend_income',
    'A01000': 'capital_gain'
}

IRS = IRS.rename(columns=irs_rename)
IRS = IRS[['state', 'zip_code', 'income_bracket', 'num_returns', 
           'total_income', 'wage_income', 'interest_income', 
           'dividend_income', 'capital_gain']]


IRS.head()
```

```python
# 重命名
real_estate_rename = {
    'brokered_by': 'broker_id',
    'status': 'listing_status',
    'bed': 'bedrooms',
    'bath': 'bathrooms',
    'acre_lot': 'lot_size',
    'street': 'street_address',
    'prev_sold_date': 'previous_sold_date'
}

real_estate = real_estate.rename(columns=real_estate_rename)

# 创建新的 property_id
real_estate.insert(0, 'property_id', range(1, len(real_estate) + 1))

# 只保留需要的列
real_estate = real_estate[['property_id', 'broker_id', 'listing_status', 'price',
                            'bedrooms', 'bathrooms', 'lot_size', 'street_address',
                            'city', 'state', 'house_size', 'zip_code', 'previous_sold_date']]
```

```python
real_estate = real_estate.reset_index(drop=True)
real_estate.dropna(inplace=True)
real_estate.head()
```

```python
Education.to_csv('../csv/Education_Final.csv', index=False)
real_estate.to_csv('../csv/Real_Estate_Final.csv', index=False)
IRS.to_csv('../csv/IRS_Final.csv', index=False)
```
